```python
from pandas import DataFrame

df= {"Sales" :[261.96,731.94,14.62,957.5775,22.368,48.86,7.28,907.152,18.504,114],
     "Profit" : [41.9136,219.582,6.8714,-383.031,2.5164,14.1694,1.9656,90.7152,5.7825,34.47]}

df = DataFrame(df,columns = ["Sales","Profit"])
df.plot(kind = "line",x = "Sales",y = "Profit")
plt.show()
```

#### Here is graph of Sales and Profit, which increases and decreases of the values ,here taken ten values in which sales and profit , can be increaes and decreases while we take all the values. 

```python
#Top 10 States with more sales
top_10_sales = df.groupby("State").Sales.sum().nlargest(10)
top_10_sales
```

```python
plt.style.use('seaborn')
top_10_sales.plot(kind= 'bar', figsize = (14,8), fontsize = 14)
plt.xlabel("States", fontsize = 13)
plt.ylabel("Total Sales", fontsize = 13)
plt.title("Top 10 States by Sales", fontsize = 16)
plt.show()
```

### California has the maximum sales compared to the other states 

```python
#Top 10 States with more profit
top_10_profit = df.groupby("State").Profit.sum().nlargest(10)
top_10_profit
```

```python
plt.style.use('seaborn')
top_10_profit.plot(kind = 'bar', figsize =(14,8), fontsize = 14)
plt.xlabel("States", fontsize = 13)
plt.ylabel("Profit", fontsize = 13)
plt.title("Total 10 States by profit", fontsize = 16)
plt.show()
```

### California and New York has maximum profit, where we can make more profit


## Here is the little comparision between sales, profit and discount 

```python
plt.style.use('seaborn')
df.plot(kind= 'scatter' , figsize = (14,8), x = 'Sales', y = 'Profit',c = 'Discount',colormap = 'viridis')
plt.ylabel('Total Profit', fontsize = 13)
plt.title('Interdependence of Sales,Profit,Discount',fontsize = 16)
plt.show()
```

#### According with the graph, California and New York has more purchases ,which means we can increase demand.

## What is the most profitable category and sub -category?

```python
#Visualzing Category section

sns.set_style("whitegrid")
sns.countplot(x = "Category", data = df, palette = "rainbow")
```

#### Most of the items are from Office supplies and less in technology

```python
#Visualzing the sub-catogory section

plt.figure(figsize = (10,10))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.2f%%')
plt.show()
```
